#### 1. Qual a  média do preço de compra dos imóveis por “Nível”?
- Nível 0: entre R$\$$ 0,00 e R$\$$ 321.950,00
- Nível 1: entre R$\$$ 321.950,00 e R$\$$ 450.000,00
- Nível 2: entre R$\$$ 450.000,00 e R$\$$ 645.000,00
- Nível 3: Preço acima de R$\$$ 645.000,00

```python
for i in range(len(data)):
    if (data.loc[i, 'price'] >= 0) & (data.loc[i, 'price'] <= 321950):
        data.loc[i, 'nivel'] = 'nivel_0'
    elif (data.loc[i, 'price'] > 321950) & (data.loc[i, 'price'] <= 450000):
        data.loc[i, 'nivel'] = 'nivel_1'
    elif (data.loc[i, 'price'] > 450000) & (data.loc[i, 'price'] <= 645000):
        data.loc[i, 'nivel'] = 'nivel_2'
    else:
        data.loc[i, 'nivel'] = 'nivel_3'
data.head()
```

```python
avg_price_by_nivel = data[['price', 'nivel']].groupby('nivel').mean().reset_index()
avg_price_by_nivel
```

#### 2. Qual a média do tamanho da sala de estar dos imóveis por "Size"?
- Size 0: Tamanho entre 0 e 1427 sqft
- Size 1: Tamanho entre 1427 e 1910 sqft
- Size 2: Tamanho entre 1910 e 2550 sqft
- Size 3: Tamanho acima de 2550 sqtf

```python
data['size'] = data['sqft_living'].apply(lambda x: 0 if x < 1427 else
                                          1 if (x > 1427) & (x < 1910) else
                                          2 if (x > 1910) & (x < 2550) else 3)
data.head()
```

```python
avg_sqft_living_by_size = data[['sqft_living', 'size']].groupby('size').mean().reset_index()
avg_sqft_living_by_size
```

#### 3. Adicione as seguintes informações ao conjunto de dados original:
- Place ID: Identificação da localização
- OSM type: Open Street Map Type
- Country: Nome do País
- Country Code: Código do País

```python
from geopy.geocoders import Nominatim
```

```python
# Initialize Nominatim API
geolocator = Nominatim(user_agent='geoapiExercises')
```

```python
# exemplo teste
response = geolocator.reverse('47.5112,-122.257')
response.raw
```

```python
# Create empty rows
data['place_id'] = 'NA'
data['osm_type'] = 'NA'
data['country'] = 'NA'
data['country_code'] = 'NA'
```

```python
geolocator = Nominatim(user_agent='geopiExercises')

for i in range(len(data)):
    if i % 1000 == 0:
        print(f'loop: {i} / {len(data)}')
    query = str(data['lat'][i]) + ',' + str(data['long'][i])
    
    #API Request
    response = geolocator.reverse(query)
    
    #populate data
    if 'place_id' in response.raw:
        data.loc[i, 'place_id'] = response.raw['place_id']
    if 'osm_type' in response.raw:
        data.loc[i, 'osm_type'] = response.raw['osm_type']
    if 'country' in response.raw['address']:
        data.loc[i, 'country'] = response.raw['address']['country']
    if 'country_code' in response.raw['address']:
        data.loc[i, 'country_code'] = response.raw['address']['country_code']
```

```python
data.to_csv('datasets/kc_house_data2.csv', index=False)
```

#### 4. Adicione os seguintes filtros no mapa:
- Tamanho mínimo da área da sala de estar
- Número mínino de banheiros
- Valor máximo do preço
- Tamanho máximo da área do porão
- Filtro das condições do imóvel
- Filtro por ano de construção

```python
import pandas as pd
import plotly.express as px
import ipywidgets as widgets
```

```python
data = pd.read_csv('datasets/kc_house_data2.csv')
```

```python
data['is_waterfront'] = data['waterfront'].apply(lambda x: 'yes' if x==1 else 'no')

data['level'] = data['price'].apply(lambda x: 0 if x < 321950 else
                                          1 if (x > 321950) & (x < 450000) else
                                          2 if (x > 450000) & (x < 645000) else 3)
data['level'] = data['level'].astype(int)
```

```python
# FILTROS #

style = {'description_width': 'initial'}

# Botão interativo de vista para o mar (dropdown)
waterfront_bar = widgets.Dropdown(
    options=data['is_waterfront'].unique().tolist(),
    value='yes',
    description='Water View',
    disable=False)

# 1. Botão interativo do tamanho da sala de estar (slider)
sqft_living_slider = widgets.IntSlider(
    value=data['sqft_living'].min(),
    min=data['sqft_living'].min(),
    max=data['sqft_living'].max(),
    step=1,
    description='Liv Room Min Size',
    dizable=False,
    style=style)

# 2. Botão interativo de número mínino de banheiros (dropdown)
bathrooms_bar = widgets.Dropdown(
    options=sorted(data['bathrooms'].unique().tolist()),
    value=0,
    description='Bathrooms',
    disable=False)

# 3. Botão interativo do preço (slider)
price_slider = widgets.FloatSlider(
    value=data['price'].max(),
    min=data['price'].min(),
    max=data['price'].max(),
    step=1,
    description='Maximum Price',
    dizable=False,
    readout_format='.2f',
    style=style)

# 4. Botão interativo do do tamanho máximo da área do porão (slider)
basement_slider = widgets.IntSlider(
    value=data['sqft_basement'].max(),
    min=data['sqft_basement'].min(),
    max=data['sqft_basement'].max(),
    step=1,
    description='Basement Size',
    dizable=False,
    style=style)

# 5. Filtro das condições do imóvel
condition_button = widgets.ToggleButtons(
    options=sorted(data['condition'].unique()),
    value=3,
    description='Condition:',
    disabled=False,
    style=style)

# 6. Filtro por ano de construção
year_slider = widgets.IntRangeSlider(
    value=[data['yr_built'].min(), data['yr_built'].max()],
    min=data['yr_built'].min(),
    max=data['yr_built'].max(),
    step=1,
    description='Year:',
    disabled=False)

def interact_map(data, waterfront, p_slider, s_sqft_liv, b_bar, base_slider, cond_button, yr_slider):
    houses = data[(data['price'] <= p_slider) &
                  (data['is_waterfront'] == waterfront) &
                  (data['sqft_living'] >= s_sqft_liv) &
                  (data['bathrooms'] >= b_bar) &
                  (data['sqft_basement'] <= base_slider) &
                  (data['condition'] == cond_button) &
                  (data['yr_built'] != yr_slider)]
    [['id', 'lat', 'long', 'price', 'level', 'sqft_living', 'bathrooms', 'sqft_basement', 'condition', 'yr_built']]

    fig = px.scatter_mapbox(houses, lat='lat', lon='long', color='level', size='price',
                           color_continuous_scale=px.colors.cyclical.IceFire,
                           size_max=15, zoom=10)

    fig.update_layout(mapbox_style='open-street-map')
    fig.update_layout(height=600, margin={'r':0, 't':0, 'l':0, 'b':0})
    fig.show()
```

```python
widgets.interactive(interact_map,
                    data=widgets.fixed(data),
                    waterfront=waterfront_bar,
                    p_slider=price_slider,
                    s_sqft_liv=sqft_living_slider,
                    b_bar=bathrooms_bar,
                    base_slider=basement_slider,
                    cond_button=condition_button, yr_slider=year_slider)
```
